```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'State': ['Maharashtra', 'Kerala', 'Delhi', 'Karnataka'],
    'Cases': [320, 250, 150, 180]
}

df = pd.DataFrame(data)

plt.figure(figsize=(8,5))
plt.hist(df['Cases'], bins=5, color='orange', edgecolor='black')
plt.title('Histogram of COVID-19 Cases Across Indian States (2020)', fontsize=14, fontweight='bold')
plt.xlabel('Cases (in thousands)', fontsize=12)
plt.ylabel('Frequency of States', fontsize=12)
plt.tight_layout()
plt.show()
```
